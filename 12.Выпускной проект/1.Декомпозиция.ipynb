# Описание проекта

Телеком провайдер «Нупозвони» нанял нас на должность джуниор аналитика. Клиенты компании -- колл-центры, мы хотим помочь им найти неэффективных операторов. Наш заказчик -- продуктовый менеджер. Он и отдел продуктовой разработки обратились к нам с просьбой выяснить сколько таких операторов и будет ли спрос у нового функционала. В дальнейшем, на основе нашего исследования, будет принято решение о разработке нового алгоритма выявления неэффективных операторов для встраивания в сервис. Выявление неэффективных операторов позволит в дальнейшем увеличить ценность сервиса для клиента, а значит потенциально увеличит время жизни клиента. Заказчику интересно какому числу клиентов данный функционал будет интересен, сколько таких неэффективных операторов и к каким клиентам стоит идти в первую очередь с данным функционалом. 

# Описание данных

Датасеты описывают использование услуг «Нупозвони» — провайдера виртуальный телефонии. Его клиенты — колл-центры, которые: 

- распределяют входящие вызовы на операторов,
- совершают исходящие вызовы силами операторов.

Также операторы могут делать внутренние вызовы -- вызовы друг между другом внутри сети виртуальной телефонии.

Колонки в `telecom_dataset.csv`:

- `user_id` — Идентификатор клиентского аккаунта в сервисе
- `date` — Дата статистики
- `direction` — Направление вызовов (out - исходящий вызов, in — входящий вызов)
- `internal` — Является ли звонок внутренним звонком между операторами клиента
- `operator_id` — Идентификатор оператора
- `is_missed_call` — Является ли звонок пропущенным
- `calls_count` — Количество звонков
- `call_duration` — Длительность звонка (без учета времени ожидания)
- `total_call_duration` — Длительность звонка (с учетом времени ожидания)

Колонки в `telecom_clients.csv`:

- `user_id` — Идентификатор клиентского аккаунта в сервисе
- `tariff_plan` — Текущий тарифный план клиента
- `date_start` — Дата регистрации клиентов в сервисе

# Инструкция по выполнению проекта

## Загрузка данных и подготовка к исследованию

Загрузить данные об использование услуг провайдера "Нупозвони". Необходимо проверить типы данных в каждой колонке, есть ли пропущенные значения и дубликаты (если такие имеются обработать их). Проверить, что в данных нет аномалий (значения, которые не отражают действительность и появились по какой-то ошибке, например отрицательное время звонка), и проверить названия колонок (если требуется привести их к удобному регистру). 

## Исследовательский анализ данных

1. Исследуем колл-центры:
- Посчитаем, сколько всего колл-центров.
- Посмотрим на распределение количества операторов по колл-центрам.
- Посмотрим на распределение количества внешних звонков по колл-центрам.
- Посмотрим, какие колл-центры чаще получают звонки при обслуживании клиентов, чем делают сами. Для этого посчитаем процент исходящих внешних звонков от общего количества внешних звонков и построим его распределение по колл-центрам.

2. Исследуем работу операторов:
- Посмотрим на распределение количества внешних звонков по операторам.
- Посмотрим на распределение процента внешних совершенных звонков по отношению ко всем внешним звонкам по операторам.
- Посмотрим распределение среднего количества/длительности внешних звонков в день по операторам в разбивке по направлению звонка.
- Посмотрим на распределение процента количества и длительности совершенных внутренних звонков к общему количеству и длительности непропущенных внешних звонков по операторам.
- Посмотрим на распределение процента количества внешних входящих пропущенных звонков по отношению к общему количеству внешних входящих звонков по каждому оператору.
- Посмотрим на распределение процента времени ожидания по внешним входящим звонкам от общей длительности внешних входящих звонков по операторам.

## Выявление неэффективных операторов 

Будем считать, что эффективность оператора зависит от функции колл-центра — принимать или совершать звонки.

1. Для колл-центра, который преимущественно совершает звонки будем считать оператора неэффективным, если: 

    * оператор совершает условно мало звонков 
    
    Можем поделить суммарную длительность внешних звонков оператора на длительность отработанных дней (кол-во смен * 8 часов). Получим процент рабочего времени, потраченного впустую. Будем считать эффективным оператора с процентом рабочего времени > 90. 

2. Для колл-центра, который преимущественно принимает звонки будем считать оператора неэффективным, если он:

    - условно долго не берет трубку (оператор в среднем заставляет клиента ждать дольше, чем готово ждать большинство клиентов)
    
    - пропускает условно много входящих звонков 

## Проверка статистических гипотез 

Гипотеза № 1: Среднее количество входящих звонков в день и среднее количество исходящих звонков в день равны.

Гипотеза № 2: Среднее количество звонков в день в принимающем колл-центре и среднее количество звонков в день в обзванивающем колл-центре равны.
